```python
#load packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load file
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

```python
data.dtypes
```

```python
data.shape
```

```python
data['Effective To Date'] = pd.to_datetime(data['Effective To Date'])
```

```python
#Load the continuous and discrete variables into continuous_df and discrete_df variables.
continuous_df = data.select_dtypes(np.number)
discrete_df = data.select_dtypes(np.object)
```

```python
#Plot a correlation matrix, what can you see?
plt.figure(figsize=(20,15))
sns.heatmap(continuous_df.corr(), annot=True)
plt.show()
```

```python
def df_bar(df):
    sns.set(rc={'figure.figsize':(16,8)})
    
    for i, col in enumerate(df):
        plt.figure(i)
        sns.barplot(x = df[col].value_counts().index, y = df[col].value_counts())
        plt.show()
```

```python
df_bar(discrete_df[['State', 'Response']])
```
